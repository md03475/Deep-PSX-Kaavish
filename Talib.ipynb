### Historic Data

```python
abl = pd.read_csv('ABL PA Equity.csv')
abl.tail()
```

### Simple Moving Average

```python
close = abl['Close']
output = talib.SMA(close, timeperiod=21)
abl['21 SMA'] = output 
abl.tail()
```

```python
close = abl['Close']
output = talib.SMA(close, timeperiod=50)
abl['50 SMA'] = output 
abl.tail()
```

### RSI

```python
real = talib.RSI(close, timeperiod=14)
abl['14 RSI'] = real 
abl.tail()
```

### Bollinger Bands

```python
upperband, middleband, lowerband = talib.BBANDS(close, timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
abl['Upper BBand'] = upperband
abl['Middle BBand'] = middleband
abl['Lower BBand'] = lowerband
abl.tail()
```

### MACD

```python
macd, macdsignal, macdhist = talib.MACD(close, fastperiod=12, slowperiod=26, signalperiod=9)
trend = []
for i in range(len(macdhist)):
    if (i <= 32):
        trend.append('NaN')
    else:
        if (macdhist[i] > 0):
            trend.append('Bullish')
        else:
            trend.append('Bearish')
        
abl['MACD'] = trend
abl.tail()
```

### 52 Week High

```python
high = []
for i in range(len(abl['Close'])):
    if (i >=260):
        temp = abl['Close'][i-260:i]
        high.append(max(temp))
    else:
        high.append('NaN')

abl['52-week high'] = high
abl.tail()
```

### 52 Week Low

```python
low = []
for i in range(len(abl['Close'])):
    if (i >=260):
        temp = abl['Close'][i-260:i]
        low.append(min(temp))
    else:
        low.append('NaN')

abl['52-week low'] = low
abl.tail()
```

### Average Volume

```python
avg = []
for i in range(len(abl['Volume'])):
    if (i >=21):
        temp = abl['Volume'][i-21:i]
        avg.append(sum(temp)/len(temp))
    else:
        avg.append('NaN')

abl['21-avg volume'] = avg
abl.tail()
```
