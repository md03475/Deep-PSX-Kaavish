```python
f = open("closing11-9Oct.lis", "r")
```

```python
import csv
with open('9Oct.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    for line in f:
        i = 0
        Date = ''
        Ticker = ''
        x=''
        y=''
        Open = ''
        High = ''
        Low = ''
        Close = ''
        Volume = ''
        while line[i]!='|':
            Date+=line[i]
            i+=1
        i+=1
    
        while line[i]!='|':
            Ticker+=line[i]
            i+=1
        i+=1
    
        while line[i]!='|':
            x+=line[i]
            i+=1
        i+=1
    
        while line[i]!='|':
            y+=line[i]
            i+=1
        i+=1
    
        while line[i]!='|':
            Open+=line[i]
            i+=1
        i+=1
    
        while line[i]!='|':
            High+=line[i]
            i+=1
        i+=1
    
        while line[i]!='|':
            Low+=line[i]
            i+=1
        i+=1
    
        while line[i]!='|':
            Close+=line[i]
            i+=1
        i+=1
    
        while line[i]!='|':
            Volume+=line[i]
            i+=1
        i+=1
    
        writer.writerow([Date, Ticker, x, y, Open, High, Low, Close, Volume])    
        
```
