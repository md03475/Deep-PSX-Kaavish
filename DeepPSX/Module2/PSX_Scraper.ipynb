```python
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.select import Select
import math  
```

```python
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.get("https://dps.psx.com.pk/announcements/companies")
```

```python
monthList = ['January','February','March','April','May','June','July','August','September','October','November','December']
yearList = ['2016','2017','2018','2019','2020','2021']
```

```python
pdf_link, year = [], []
```

```python
for currentYear in yearList:
  yearSearch = driver.find_element_by_xpath("//select[@name='sector']/option[text()='" + currentYear + "']").click()
  print(currentYear)
  for currentMonth in monthList:
    print(currentMonth)
    monthSearch = driver.find_element_by_xpath("//select[@name='sector']/option[text()='" + currentMonth + "']").click()
    searchButton = driver.find_element_by_xpath('//button[text()="SEARCH"]')
    searchButton.click()
    time.sleep(25)
    number = driver.find_element_by_class_name('announcementsResults__header')
    extract = number.text[19:23]
    print(extract)
    pages = math.ceil(int(extract) / 50)
    print(pages)
    for x in range (1,pages+1):
      try:
        for i in range (1,51):
          l = driver.find_elements_by_xpath ("//*[@class= 'tbl']/tbody/tr["+str(i)+"]/td")
          while (len(l) != 0):
            try:
              if "PDF" in l[5].text:
                link = l[5].find_element_by_partial_link_text("PDF")
                pdf_link.append(link.get_attribute('href'))
                year.append(l[0].text)
                l=[]
                #download_file(pdf_link[i-1], year[i-1] + '.'*i)
              else:
                pdf_link.append(i)
                year.append(i)
                l=[]
            except StaleElementReferenceException:
              l=[]
              continue
        
        next = driver.find_element_by_xpath('//button[text()="Next"]')
        next.click()
        time.sleep(10)
      except NoSuchElementException:
        continue
  print(pdf_link)
  print(year)
```

```python
import pandas as pd
df = pd.DataFrame()
df['Date'] = [i for i in year if type(i) != int]
df['Link'] = [i for i in pdf_link if type(i) != int]
df.head()
```
